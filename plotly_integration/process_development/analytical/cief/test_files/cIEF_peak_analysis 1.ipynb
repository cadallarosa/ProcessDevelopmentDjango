Highest peak area

```python
Peak_Area = []
for path in output_paths:
    df = pd.read_csv(path, header=0)
    df = df[(df['Time']>LB) & (df['Time']<HB)]
    df['Area %'] = df['Area']/sum(df['Area'])*100
    Peak_Area = np.append(Peak_Area, df['Area %'][df['Height']==df['Height'].max()].values)
    print(df)
    
print(Peak_Area)
```

Sum of the peaks before highest area 

```python
BASE = []
for path in output_paths:
    df = pd.read_csv(path, header=0)
    df = df[(df['Time']>LB) & (df['Time']<HB)]
    df['Area %'] = df['Area']/sum(df['Area'])*100
    df.index = range(1, len(df.index)+1)
    #Bool = (df['Area %'] == df['Area %'].max())
    Bool = df['Height']==df['Height'].max()
    j = Bool[Bool==True].last_valid_index()
    for i in range(1, len(Bool)+1):
        if i < j:
            Bool[i] = True
        elif i == Bool[Bool==True].last_valid_index():
            Bool[i] = False
    BASE += [sum(df['Area %'][Bool])]
    print(df)
    
print(BASE)
```

Sum of the peaks after the peak area

```python
##sum of the peaks after peak area
ACID = []
for path in output_paths:
    df = pd.read_csv(path, header=0)
    df = df[(df['Time']>LB) & (df['Time']<HB)]
    df['Area %'] = df['Area']/sum(df['Area'])*100
    
    df.index = range(1, len(df.index)+1)
    #Bool = (df['Area %'] == df['Area %'].max())
    Bool = df['Height']==df['Height'].max()
    j = Bool[Bool==True].last_valid_index().astype(int)
    for i in range(1, len(Bool)+1):
        if  i > j:
            Bool[i] = True
        elif i == Bool[Bool==True].last_valid_index():
            Bool[i] = False
    ACID += [sum(df['Area %'][Bool])]
    print(df)
    
print(ACID)
```

```python
ACID = np.round(ACID,2)
BASE = np.round(BASE,2)
Peak_Area = np.round(Peak_Area,2)
df = pd.DataFrame([ACID, Peak_Area, BASE], index = ("ACID", "Peak_Area", "BASE"), columns = output_paths)
df.to_csv("cIEF_Peak_Summary.csv")
print(df)
```
